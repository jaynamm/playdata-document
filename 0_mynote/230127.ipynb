```python
import os
import shutil

base_dir = "/Users/jaynam/workspace/playdata/이름_생년_성별_10000/"
target_dir = "/Users/jaynam/workspace/playdata/people/"

for f in file_list:
    print(f)
    gender_num = f.split("_")[2][0]
    print(gender_num)
    
    if gender_num == '1' or gender_num == '3':
        shutil.move(base_dir + f, target_dir + "male/" + f)
    elif gender_num == '2' or gender_num == '4':
        shutil.move(base_dir + f, target_dir + "female/" + f)
```

```python
import os

male_dir = "/Users/jaynam/workspace/playdata/people/male/"
female_dir = "/Users/jaynam/workspace/playdata/people/female/"

md = []

for files in os.walk(male_dir):
    mfile = [f for f in files[2]]
    
        
for f in mfile:
    with open(male_dir + f, "r") as mf:
        md += [int(mf.read().replace(",", ""))]

        
print("[남자] 합계 : ", sum(md), " 평균 : ", sum(md)/len(md))
        
    
fd = []
        
for files in os.walk(female_dir):
    ffile = [f for f in files[2]]


for f in ffile:
    with open(female_dir + f, "r") as ff:
        fd += [int(ff.read().replace(",", ""))]

        
print("[여자] 합계 : ", sum(fd), " 평균 : ", sum(fd)/len(fd))
```

```python
import selenium
from selenium.webdriver.common.by import By

driver = selenium.webdriver.Chrome()

url = "https://www.koreabaseball.com/"

driver.get(url)

driver.find_element(By.CSS_SELECTOR,"#popupVideo > img").click()                
driver.find_element(By.CSS_SELECTOR,"#lnb > li:nth-child(4) > a").click()
```

```python
import time
import re

select_team = "#cphContents_cphContents_cphContents_ddlTeam > option:nth-child({})"
select_page = "#cphContents_cphContents_cphContents_ucPager_btnNo{}"
pattern = re.compile("playerId=([0-9]+)")

playid = []

for x in range(2,12):
    for_1 = select_team.format(x)
    driver.find_element(By.CSS_SELECTOR, for_1).click()
    time.sleep(2)
    #playid.extend(pattern.findall(driver.page_source))
    for y in range(1,6):
        f2 = select_page.format(y)
        try:
            driver.find_element(By.CSS_SELECTOR, f2).click()
            time.sleep(1)
            playid.extend(pattern.findall(driver.page_source))
        except Exception as e:
            print (e)
        time.sleep(2)
```
