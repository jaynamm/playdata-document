```python
def fib(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b
        
# n = 0 -> a = 0     / b = 1         -> a = 0
# n = 1 -> a = b = 1 / b = a + b = 1 -> a = 1
# n = 2 -> a = b = 1 / b = a + b = 2 -> a = 1
# n = 3 -> a = b = 2 / b = a + b = 3 -> a = 2
```
