# 데이터프레임 합치기

## [pd.merge()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.merge.html)

```python
import pandas as pd 
```

```python
fruit = pd.DataFrame({'Num':[123, 456, 789, 1011, 1112], 'Fruit':['Apple', 'Banana', 'Cherry', 'Lemon', 'Peach']})
grade = pd.DataFrame({'Num':[123, 789, 1314], 'Grade':['A', 'B', 'C']})

print(fruit, '\n')
print(grade)
```

### Left Join

```python
pd.merge(fruit, grade, on = 'Num', how = 'left')
```

```python
pd.merge(grade, fruit, on = 'Num', how = 'left') # pd.merge(fruit, grade, on = 'Num', how = 'right')
```

### Inner Join

```python
pd.merge(fruit, grade, on = 'Num', how = 'inner')
```

### Outer Join

```python
pd.merge(fruit, grade, on = 'Num', how = 'outer')
```

## [pd.concat()](https://pandas.pydata.org/docs/reference/api/pandas.concat.html)
> 물리적으로(인덱스 또는 컬럼) 이어 붙여주는 함수

```python
import pandas as pd 
```

```python
df1 = pd.DataFrame({'a':['a0','a1','a2','a3'],
                    'b':['b0','b1','b2','b3'],
                    'c':['c0','c1','c2','c3']},
                    index = [0,1,2,3])

df2 = pd.DataFrame({'a':['a2','a3','a4','a5'],
                    'b':['b2','b3','b4','b5'],
                    'c':['c2','c3','c4','c5'],
                    'd':['d2','d3','d4','d5']},
                    index = [2,3,4,5])

print(df1, '\n')
print(df2)
```

### 행방향으로 합치기

아래 결과를 보면,     
위 아래 방향으로 합치기 & 하지만 인덱스 값에 중복이 있음

```python
result1 = pd.concat([df1,df2], axis=0)
result1 
```

인덱스 재정의

```python
result2 = pd.concat([df1,df2], axis=0)
result2.reset_index(inplace=True)
result2 
```

inner join

```python
result3 = pd.concat([df1,df2], ignore_index=True, join='inner')
result3 
```

### 열방향으로 합치기

```python
result4 = pd.concat([df1,df2],axis=1)
result4 
```

inner join

```python
result5 = pd.concat([df1,df2], axis=1, join='inner')
result5  
```

인덱스 재정의

```python
result5.reset_index(inplace=True)
result5
```

### 시리즈와 합치기

```python
sr1 = pd.Series(['e0','e1','e2','e3'], name = 'e')
sr1 
```

```python
result6 = pd.concat([df1,sr1], axis=1)
result6
```
