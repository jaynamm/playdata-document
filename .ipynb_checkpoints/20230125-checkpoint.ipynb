```python
import requests
from bs4 import BeautifulSoup

headers = {
    "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/95.0.4638.69 Safari/537.36"
}

ids = [songid['data-song-no'] for songid in BeautifulSoup(r.text).find('table').find_all("tr", id='lst50')] + [songid['data-song-no'] for songid in BeautifulSoup(r.text).find('table').find_all("tr", id='lst100')]
titles = [songid.find('a', 'song_info')['title'].replace("곡정보", "").strip() for songid in BeautifulSoup(r.text).find('table').find_all("tr", id='lst50')] + [songid.find('a', 'song_info')['title'].replace("곡정보", "").strip() for songid in BeautifulSoup(r.text).find('table').find_all("tr", id='lst100')]
artists = [songid.find('div', 'rank02').find('a').text for songid in BeautifulSoup(r.text).find('table').find_all("tr", id='lst50')] + [songid.find('div', 'rank02').find('a').text for songid in BeautifulSoup(r.text).find('table').find_all("tr", id='lst100')]

total = [[ids[i], titles[i], artists[i]] for i in range(100)]

# for id in total_id:
#     print(f"https://www.melon.com/song/detail.htm?songId={id}")

# song_url = "https://www.melon.com/song/detail.htm?songId=35945927"

for song in total:
    song_url = f"https://www.melon.com/song/detail.htm?songId={song[0]}"

    get_song = requests.get(song_url, headers = headers)

    bs = BeautifulSoup(get_song.text).find('div', id='d_video_summary')

    song_lyrics = BeautifulSoup(str(bs).replace('<br/>', '\n')).text.strip()

    f = open(f"/Users/jaynam/workspace/playdata/song/{song[0]} - {song[1]} - {song[2]}", "w", encoding = "utf-8")
    f.write(song_lyrics)
    f.close()
```
